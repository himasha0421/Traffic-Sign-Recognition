### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.
import numpy as np
# 1. normalize the dataset min-max scaling
def normalize_image(Dataset):
    min_range=0.1
    max_range=0.9
    normalize_images = []
    #take the mean of the image across color channels
    
    for idx in range(len(Dataset)):
        image=np.array(Dataset[idx])
        mean_img = np.mean(image , axis=2)
        x_max = max(mean_img.reshape((-1,1)))
        x_min = min(mean_img.reshape((-1,1)))
        img_flat = min_range + ((mean_img.reshape((-1,1)) - x_min)/(x_max - x_min))*(max_range - min_range)
        img_resize = img_flat.reshape((32,32,1))
        normalize_images.append(img_resize)
        
    return normalize_images
    
````

````python
train_features = normalize_image(X_train)
test_features = normalize_image(X_test)
valid_features = normalize_image(X_valid)
````

### Model Architecture

### LeNet5 Model

````python
import tensorflow as tf
#define LeNet5

def LeNet5(x , keep_prob):
    mu=0
    sigma = 0.1
    #conv 1 input_shape(32,32,1) output (28,28,6)
    Wc1 = tf.Variable(tf.random_normal([5,5,1,6] , mean=mu , stddev = sigma))
    bc1 = tf.Variable(tf.zeros([6]))
    conv_1 = tf.nn.conv2d(x , Wc1 ,  strides=[1,1,1,1] , padding="VALID")
    conv_1 = tf.nn.bias_add(conv_1 , bc1)
    #relu activation
    conv_1_relu = tf.nn.relu(conv_1 , name='conv_1')
    conv_1_max = tf.nn.max_pool(conv_1_relu , ksize=[1,2,2,1],strides=[1,2,2,1],padding="SAME")
    
    #conv_2 input (14,14,6) output (10,10,16)
    Wc2 = tf.Variable(tf.random_normal([5,5,6,16],mean=mu , stddev = sigma))
    bc2 = tf.Variable(tf.zeros([16]))
    conv_2 = tf.nn.conv2d(conv_1_max , Wc2 , strides=[1,1,1,1] , padding="VALID")
    conv_2 =  tf.nn.bias_add(conv_2 , bc2)
    #relu activation
    conv_2_relu = tf.nn.relu(conv_2 , name = "conv_2")
    conv_2_max = tf.nn.max_pool(conv_2_relu , ksize=[1,2,2,1] , strides=[1,2,2,1],padding="SAME")
    
    #flatten layer
    Dense_1 = flatten(conv_2_max)
    #dense 1 input 400 output 120
    Wd1 = tf.Variable(tf.random_normal([400,120],mean=mu , stddev = sigma))
    bd1 = tf.Variable(tf.zeros([120]))
    Dense_2 =tf.add(tf.matmul(Dense_1  , Wd1),bd1)
    Dense_2_relu = tf.nn.relu(Dense_2)
    Dense_2_drop = tf.nn.dropout(Dense_2_relu,keep_prob=keep_prob)
    
    #dense 2 input 120 output 84
    Wd2 = tf.Variable(tf.random_normal([120,84],mean=mu , stddev = sigma))
    bd2 = tf.Variable(tf.zeros([84]))
    Dense_3 = tf.add(tf.matmul(Dense_2_drop,Wd2),bd2)
    Dense_3_relu = tf.nn.relu(Dense_3)
    Dense_3_dropout = tf.nn.dropout(Dense_3_relu,keep_prob=keep_prob)
    
    #dense 3 input 84 output 43
    Wd3 = tf.Variable(tf.random_normal([84,43],mean=mu , stddev= sigma))
    bd3 = tf.Variable(tf.zeros([43]))
    logits = tf.add(tf.matmul(Dense_3_dropout,Wd3),bd3)
    
    return logits
````

````python
#define the place holders from featuers , labels , keep_prob
features = tf.placeholder(tf.float32 , [None , 32 , 32 ,1])
y_value = tf.placeholder(tf.int32 , [None])
labels =tf.one_hot(y_value , n_classes)
keep_prob = tf.placeholder(tf.float32)
logits = LeNet5(features ,keep_prob)
predictions = tf.nn.softmax(logits)
learning_rate=0.01
batch_size=128
#define the cost function and optimizer
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits = logits , labels =labels))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
training_operation = optimizer.minimize(cost)
````

````python
#accuracy functions
correct_labels=tf.equal(tf.argmax(labels,1),tf.argmax(predictions,1))
accuracy = tf.reduce_mean(tf.cast(correct_labels , tf.float32))
````

````python
train_dict={'features':train_features , 'labels':y_train}
valid_dict = {'features':valid_features , 'labels':y_valid}
test_dict ={'features':test_features , 'labels':y_test}
````

````python
def evaluate(x_data , y_data):
    data_size= len(x_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for idx in range(0,data_size, batch_size):
        batch_x , batch_y=x_data[idx:idx+batch_size], y_data[idx:idx+batch_size]
        valid_acc = sess.run(accuracy , feed_dict={features:batch_x , y_value:batch_y , keep_prob:1.0})
        total_accuracy+=valid_acc*len(batch_x)
    valid_accuracy = total_accuracy/data_size
    
    return valid_accuracy
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
from sklearn.utils import shuffle
n_epochs = 20
log_batch_size=50
batches=[]
loss_batch=[]
train_acc_batch=[]
valid_acc_batch=[]
validation_accuracy = 0.85
saver=tf.train.Saver()
model_path="model_lenet_final.ckpt"
#initalize the session
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(n_epochs):
        x_train , y_train = shuffle(train_dict['features'],train_dict['labels'])
        
        for offset in range(0,n_train,batch_size):
            end= offset+batch_size
            batch_x  , batch_y = x_train[offset:end],y_train[offset:end]
            _,l=sess.run([training_operation,cost] , feed_dict={features:batch_x , y_value:batch_y , keep_prob:0.7})
            
            if(offset % log_batch_size ==0):
                train_acc = sess.run(accuracy , feed_dict={features:train_dict['features'],y_value:train_dict['labels'],keep_prob:1.0})
                valid_acc = sess.run(accuracy , feed_dict={features:valid_dict['features'],y_value:valid_dict['labels'],keep_prob:1.0})
                
                previous_batch = batches[-1] if batches else 0
                batches.append(previous_batch+ log_batch_size)
                loss_batch.append(l)
                train_acc_batch.append(train_acc)
                valid_acc_batch.append(valid_acc)
                
                if(validation_accuracy < valid_acc):
                    model_path = "model_lenet_"+str(valid_acc)+"_.ckpt"
                    saver.save(sess , model_path)
        valid_accuracy = evaluate(valid_dict['features'], valid_dict['labels'])
        print("EPOCH : {} .. ".format(epoch+1))
        print("Validdation Accuracy : {:.3f}".format(valid_accuracy))
    saver.save(sess,model_path)
````

````python
loss_plot = plt.subplot(211)
loss_plot.set_title('Loss')
loss_plot.plot(batches, loss_batch, 'g')
loss_plot.set_xlim([batches[0], batches[-1]])
acc_plot = plt.subplot(212)
acc_plot.set_title('Accuracy')
acc_plot.plot(batches, train_acc_batch, 'r', label='Training Accuracy')
acc_plot.plot(batches, valid_acc_batch, 'x', label='Validation Accuracy')
acc_plot.set_ylim([0, 1.0])
acc_plot.set_xlim([batches[0], batches[-1]])
acc_plot.legend(loc=4)
plt.tight_layout()
plt.show()
````
